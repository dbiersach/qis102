```python
# eigenvalues.ipynb

# Cell 1 - Create two matrices A and B
# where matrix B is Hermitian

import numpy as np
from qis102_utils import as_latex

a = np.array(
    (
        [
            [-5.664 - 3.623j, 7.672 - 4.470j, 1.864 - 7.149j],
            [0.766 - 4.821j, -4.410 - 0.228j, 9.759 + 4.256j],
            [1.0335 - 3.672j, 3.890 - 5.741j, 7.760 + 3.812j],
        ]
    )
)


b = np.array([[5, 4 + 5j, 6 - 16j], [4 - 5j, 13, 7], [6 + 16j, 7, -2.1]])

display(as_latex(a, prefix=r"\mathbf{A}="))
display(as_latex(b, prefix=r"\mathbf{B}="))
```

```python
# Cell 2 - Eigenequation - The dot product of a matrix
# with one of its eigenvectors equals the dot product
# of the eigenvalue corresponding to that eigenvector

from IPython.core.display import Math

display(as_latex(a, prefix=r"\mathbf{A}="))

# Find the eigenvalues and eigenvectors of matrix A
eigen_vals, eigen_vecs = np.linalg.eig(a)

# Note: \mathrm is LaTeX command for Roman typeface
display(
    as_latex(eigen_vecs, prefix=r"\mathrm{Eigenvectors\;(\mathbf{v})\;of\;\mathbf{A}}=")
)
display(
    as_latex(eigen_vals, prefix=r"\mathrm{Eigenvalues\;(\lambda)\;of\;\mathbf{A}}=")
)

# Display the 2nd eigenvector and 2nd eigenvalue
display(as_latex(eigen_vecs[:, 1], prefix=r"\mathbf{v_2}="))
display(as_latex(eigen_vals[1, np.newaxis], prefix=r"\mathrm{\lambda_2}="))

t1 = np.dot(a, eigen_vecs[:, 1])
t2 = np.dot(eigen_vals[1], eigen_vecs[:, 1])

display(as_latex(t1, prefix=r"\mathbf{A\cdot v_2}="))
display(as_latex(t2, prefix=r"\mathrm{\lambda_2\cdot\mathbf{v_2}}="))

display(
    Math(
        r"\mathbf{A\cdot v_2}="
        r"\mathrm{\lambda_2\cdot\mathbf{v_2}}"
        rf"\;?\;\rightarrow\;{np.isclose(t1, t2).all()}"
    )
)
```

```python
# Cell 3 - Hermitian matrices have all real eigenvalues

display(as_latex(b, prefix=r"\mathbf{B}="))

eigen_vals, eigen_vecs = np.linalg.eig(b)

display(
    as_latex(eigen_vecs, prefix=r"\mathrm{Eigenvectors\;(\mathbf{v})\;of\;}\mathbf{B}=")
)
display(
    as_latex(eigen_vals, prefix=r"\mathrm{Eigenvalues\;(\lambda)\;of\;}\mathbf{B}=")
)
```

```python
# Cell 4 - All eigenvectors of a Hermitian matrix
# are pairwise orthogonal

display(as_latex(b, prefix=r"\mathbf{B}="))

display(as_latex(eigen_vecs[:, 1], prefix=r"\mathbf{v_1}="))
display(as_latex(eigen_vecs[:, 1], prefix=r"\mathbf{v_2}="))
display(as_latex(eigen_vecs[:, 2], prefix=r"\mathbf{v_3}="))

v1_dot_v2 = complex(np.vdot(eigen_vecs[:, 0], eigen_vecs[:, 1]))
v2_dot_v3 = complex(np.vdot(eigen_vecs[:, 1], eigen_vecs[:, 2]))
v3_dot_v1 = complex(np.vdot(eigen_vecs[:, 2], eigen_vecs[:, 0]))

display(
    Math(rf"\langle\mathbf{{v_1,v_2}}\rangle={np.round(v1_dot_v2,5)}"),
    Math(rf"\langle\mathbf{{v_2,v_3}}\rangle={np.round(v2_dot_v3,5)}"),
    Math(rf"\langle\mathbf{{v_3,v_1}}\rangle={np.round(v3_dot_v1,5)}"),
)
```

```python
# Cell 5 - Demonstrate Eigendecomposition using Diagonalizable Matrices

a = np.array(
    (
        [
            [-5.96377 + 0.563478j, -4.65625 - 0.657598j, 1.1037 + 4.11872j],
            [6.01795 - 3.17565j, -1.66041 - 9.61168j, 5.88324 + 1.86256j],
            [0.39357 - 2.25232j, -1.66388 - 15.6493j, 8.02418 + 9.0482j],
        ]
    )
)

display(as_latex(a, prefix=r"\mathbf{A}="))

eigen_vals, eigen_vecs = np.linalg.eig(a)

# The eigenvectors form the eigenbasis for a matrix
b = eigen_vecs.copy()

display(as_latex(b, prefix=r"\mathrm{Eigenbasis\;(\mathfrak{B})\;of\;}\mathbf{A}="))

# Form the transition matrix to take A to its eigenbasis
m = np.dot(np.linalg.inv(b), a)

display(
    as_latex(
        m,
        prefix=r"\mathrm{Transition\;Matrix\;of\;}(\mathbf{\mathfrak{B}\leftarrow A})=",
    )
)

# The dot product of the transition matrix (that takes a matrix to its eigenbasis)
# with its eigenbasis produces a diagonal matrix of its eigenvalues
t1 = np.dot(m, b)

# Compare to element-wise multiplication of eigenvalue vector and identity matrix
t2 = eigen_vals[np.newaxis] * np.identity(3)

display(
    as_latex(
        t1,
        places=3,
        prefix=(
            r"\mathrm{Diagonal\;Matrix\;(\mathbf{\color{red}{\Lambda}})\;=\;}"
            r"\mathbf{(\mathfrak{B}\leftarrow A)\cdot \mathfrak{B}}="
        ),
    )
)

display(as_latex(t2, places=3, prefix=r"\mathbf{\lambda\;*\;\;I}="))

display(
    Math(
        r"\mathbf{(\mathfrak{B}\leftarrow A)\cdot \mathfrak{B}}="
        r"\mathbf{\lambda\;*\;\;I}"
        rf"\;?\;\rightarrow\;{np.isclose(t1, t2).all()}"
    )
)
```

```python
# Cell 6 - Demonstrate Matrix Integer Exponentiation
# (using the more efficient diagonal matrix of eigenvalues)

a = np.array([[0.8, 0.3], [0.2, 0.7]])
display(as_latex(a, prefix=r"\mathbf{A}="))

eigen_vals, eigen_vecs = np.linalg.eig(a)

# The eigenvectors form the eigenbasis for a matrix
b = eigen_vecs.copy()
display(as_latex(b, prefix=r"\mathrm{Eigenbasis\;(\mathfrak{B})\;of\;}\mathbf{A}="))

# Create a diagonal matrix of eigenvalues
lam = eigen_vals[np.newaxis] * np.identity(2)
display(as_latex(lam, prefix=r"\mathbf{\color{#66FF66}{\Lambda}}="))

# Raise the diagonal matrix to the power of 100
pow_lam = np.linalg.matrix_power(lam, 100)
display(as_latex(pow_lam, prefix=r"\mathbf{\color{#66FF66}{\Lambda^{100}}}="))

# Raise A to the 100th using 100 array multiplications of A to itself (t1)
# and again using the more efficient eigendecomposition (t2)
t1 = np.linalg.matrix_power(a, 100)
t2 = np.dot(b, np.dot(pow_lam, np.linalg.inv(b)))

display(as_latex(t1, prefix=r"t_1=\mathbf{A^{100}}="))
display(
    as_latex(
        t2,
        prefix=(
            r"t_2="
            r"\mathbf{\mathfrak{B}\cdot\color{#66FF66}{\Lambda^{100}}}"
            r"\color{defaultcolor}{\;\cdot\;\mathfrak{B^{-1}}}"
        ),
    )
)

display(
    Math(
        r"\mathbf{A^{100}}="
        r"\mathbf{\mathfrak{B}\cdot\color{#66FF66}{\Lambda^{100}}}"
        r"\color{defaultcolor}{\;\cdot\;\mathfrak{B^{-1}}}"
        rf"\;?\;\rightarrow\;{np.isclose(t1, t2).all()}"
    )
)
```

```python
# Cell 7 - Demonstrate Matrix Commutation
# If two matrices share the same eigenbasis,
# then they will commute under matrix multiplication

# Define four arbitrary & distinct eigenvectors
eigen_vec1 = np.array([-2.7 + 5.1j, 7.8 + 2j, -11, 9j])
eigen_vec2 = np.array([1.1j, -6.2 + 0.3j, 2j, -0.7])
eigen_vec3 = np.array([8, -2.4j, 1 + 1j, -4.2 - 8.1j])
eigen_vec4 = np.array([10 - 1j, 5.5, 9.1 + 1.9j, 0])

# Create the eigenbasis from these eigenvectors
eigen_basis = np.array([eigen_vec1, eigen_vec2, eigen_vec3, eigen_vec4]).T

display(as_latex(eigen_basis, prefix=r"\mathrm{Eigenbasis\;\mathfrak{B}}="))

# Specify arbitrary & distinct eigenvalues for matrix A and matrix B
eigen_valsA = np.array([2.9 - 4.3j, 4 + 0.9j, -7j, 3.9 - 10.2j])
eigen_valsB = np.array([5.7 - 2.3j, -4.1 + 2j, 8 - 7.1j, 0.3])

display(
    as_latex(eigen_valsA, prefix=r"\mathrm{Eigenvalues\;(\lambda)\;of\;}\mathbf{A}=")
)
display(
    as_latex(eigen_valsB, prefix=r"\mathrm{Eigenvalues\;(\lambda)\;of\;}\mathbf{B}=")
)

# Diagonalize the eigenbasis (capital lambda matrices in slide deck)
lamA = eigen_valsA[np.newaxis] * np.identity(4)
lamB = eigen_valsB[np.newaxis] * np.identity(4)

# Use the eigendecomposition formula to generate matrix A and matrix B
a = np.dot(np.linalg.inv(eigen_basis), np.dot(lamA, eigen_basis))
b = np.dot(np.linalg.inv(eigen_basis), np.dot(lamB, eigen_basis))

display(as_latex(a, prefix=r"\mathbf{A}="))
display(as_latex(b, prefix=r"\mathbf{B}="))

# Test if matrix A and matrix B commute under multiplication
# Note: Matrix multiplication is generally NOT commutative,
# but is commutative if both matrices shared the same eigenbasis
t1 = np.dot(a, b)
t2 = np.dot(b, a)

display(
    Math(
        r"\mathbf{A\cdot B}="
        r"\mathbf{B\cdot A}="
        rf"\;?\;\rightarrow\;{np.isclose(t1, t2).all()}"
    )
)
```

```python
# Cell 8 - The determinant and eigenvalues of a Unitary matrix have norm 1

# Generate a random 4x4 complex matrix
n = 4
a = np.random.randn(n, n) + 1j * np.random.randn(n, n)

# Perform QR decomposition to generate the unitary matrix
u, _ = np.linalg.qr(a)

display(as_latex(a, prefix=r"\mathbf{U}="))

# Check that matrix U is unitary
# A Unitary matrix dotted with its dagger is the Identity matrix
t1 = np.dot(u, u.conj().T)
t2 = np.identity(n)

display(
    Math(rf"\mathbf{{U}}\;\mathrm{{unitary}}?\;\rightarrow\;{np.isclose(t1, t2).all()}")
)

# Compute the absolute value of the determinant
det = np.array([np.linalg.det(u)])

display(as_latex(det, prefix=r"\mathbf{det(U)}="))
display(as_latex(np.abs(det), prefix=r"\mathbf{|det(U)|}="))

# Compute the eigenvalues
eigen_vals = np.linalg.eigvals(u)

display(
    as_latex(eigen_vals, prefix=r"\mathrm{Eigenvalues\;(\lambda)\;of\;}\mathbf{U}=")
)

display(
    as_latex(
        np.abs(eigen_vals),
        prefix=r"\mathrm{|Eigenvalues|\;(\lambda)|\;of\;}\mathbf{U}=",
    )
)
```
